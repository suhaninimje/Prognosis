```python
import os
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import json
from sentinelhub import SHConfig, BBox, CRS, SentinelHubCatalog, SentinelHubDownloadClient, DataCollection, Geometry, SentinelHubRequest, MimeType, bbox_to_dimensions
from datetime import date, timedelta
from shapely.geometry import box
import pandas as pd
from concurrent.futures import ThreadPoolExecutor, as_completed
from epiweeks import Week
import folium
import rasterio

evalscript = """
//VERSION=3
function setup() {
  return {
    input: [{
      bands: ["B01", "B02", "B03", "B04", "B05", "B06", "B07", "B08", "B8A", "B09", "B11", "B12"],
      units: "DN"
    }],
    output: {
      id: "default",
      bands: 12,
      sampleType: SampleType.UINT16
    }
  }
}
function evaluatePixel(sample) {
  return [sample.B01, sample.B02, sample.B03, sample.B04,
          sample.B05, sample.B06, sample.B07, sample.B08,
          sample.B8A, sample.B09, sample.B11, sample.B12];
}
"""
```

```python
CLIENT_ID = ""
CLIENT_SECRET = ""

config = SHConfig()
config.sh_client_id = CLIENT_ID
config.sh_client_secret = CLIENT_SECRET
```

```python
total_bbox = [74.163195, 16.615629, 74.322853, 16.774081]
kolhapur_bbox = BBox(bbox=total_bbox, crs=CRS.WGS84)
geometry = Geometry({"type": "Polygon", "coordinates": [[[74.322853, 16.774081], [74.163195, 16.774081], [74.163195, 16.615629], [74.322853, 16.615629], [74.322853, 16.774081]]]}, CRS.WGS84)

# STEP 2.5: Visualize defined bounding box on interactive map
m = folium.Map(location=[(total_bbox[1]+total_bbox[3])/2, (total_bbox[0]+total_bbox[2])/2], zoom_start=10)
folium.Rectangle(
    bounds=[[total_bbox[1], total_bbox[0]], [total_bbox[3], total_bbox[2]]],
    color='red', fill=True, fill_opacity=0.3
).add_to(m)
display(m)
```

```python
start_week = Week(2018, 1)
end_week = Week(2019, 52)
weekly_intervals = [(Week(year, week).startdate(), Week(year, week).enddate())
                    for year in range(2018, 2020)
                    for week in range(1, 53)]
```

```python
catalog = SentinelHubCatalog(config=config)
output_folder = "Kolhapur_Sentinel_Images"
os.makedirs(output_folder, exist_ok=True)

results = []
def download_week_image(start, end):
    filter_str = f"eo:cloud_cover <= 100"
    search_iterator = catalog.search(
        DataCollection.SENTINEL2_L2A,
        geometry=geometry,
        time=(start, end),
        filter=filter_str,
        fields={"include": ["id", "geometry", "properties.datetime", "properties.eo:cloud_cover"]}
    )
    week_results = list(search_iterator)
    if week_results:
        best = sorted(week_results, key=lambda x: x['properties']['eo:cloud_cover'])[0]
        date_str = best['properties']['datetime'].split('T')[0]

        epiweek_str = f"{Week.fromdate(start).year}_W{str(Week.fromdate(start).week).zfill(2)}"
        image_path = os.path.join(output_folder, f"{epiweek_str}_{date_str}.tiff")
        bbox = BBox(bbox=total_bbox, crs=CRS.WGS84)
        size = bbox_to_dimensions(bbox, resolution=10)

        request = SentinelHubRequest(
            evalscript=evalscript,
            input_data=[
                SentinelHubRequest.input_data(
                    data_collection=DataCollection.SENTINEL2_L2A,
                    time_interval=(date_str, date_str),
                    mosaicking_order='leastCC'
                )
            ],
            responses=[
                SentinelHubRequest.output_response('default', MimeType.TIFF)
            ],
            bbox=bbox,
            size=size,
            config=config,
            data_folder=output_folder
        )
        try:
            temp_download_folder = os.path.join(output_folder, f"temp_{epiweek_str}_{date_str}")
            request.data_folder = temp_download_folder
            request.save_data()
            downloaded_file = os.path.join(output_folder, 'response.tiff')
            downloaded_file = os.path.join(temp_download_folder, 'response.tiff')
            if os.path.exists(downloaded_file):
                os.rename(downloaded_file, image_path)
                import shutil
                shutil.rmtree(temp_download_folder)
            results.append({
                    "epiyear": Week.fromdate(start).year,
                    "epiweek": Week.fromdate(start).week,
                    "week_start": str(start),
                    "week_end": str(end),
                    "selected_date": date_str,
                    "cloud_cover": best['properties']['eo:cloud_cover'],
                    "filename": f"{epiweek_str}_{date_str}.tiff"
                })
            print(f"✅ Downloaded for epiweek {Week.fromdate(start)}: {image_path}")
        except Exception as e:
            print(f"⚠️ Failed for epiweek {Week.fromdate(start)}: {e}")
    else:
        print(f"❌ No image available for week {Week.fromdate(start)}")

# Run direct download loop
from concurrent.futures import ThreadPoolExecutor

def should_download(start):
    epiweek_str = f"{Week.fromdate(start).year}_W{str(Week.fromdate(start).week).zfill(2)}"
    # Dummy name used for checking existence
    expected_prefix = f"{epiweek_str}_"
    return not any(f.startswith(expected_prefix) for f in os.listdir(output_folder))

with ThreadPoolExecutor(max_workers=2) as executor:
    futures = [executor.submit(download_week_image, start, end)
               for start, end in weekly_intervals if should_download(start)]
    for future in as_completed(futures):
        _ = future.result()
```

```python
results_df = pd.DataFrame(results)
results_df.to_csv("best_images_per_week_kolhapur.csv", index=False)
print("✅ Metadata CSV with epiweeks saved.")
```
